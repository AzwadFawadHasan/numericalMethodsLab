```python
# import required packages
import numpy as np
import matplotlib.pyplot as plt

a = np.array([[1,2,3],[3,4,3], [9,8,1]])
def inv(a):
    
    # Determining  the determinant of a
    det = np.linalg.det(a)
    if det == 0:
        print(" matrix has no inverse as det =0")

    # Determining the  the adjoint  matrix of a
    cfactorAll = np.zeros_like(a)
    for l in range(a.shape[0]):
        for m in range(a.shape[1]):
            minor = a[np.arange(a.shape[0]) != l][:,np.arange(a.shape[1]) != m]
            cfactor = (-1) ** (l+m) * np.linalg.det(minor)
            cfactorAll[l,m] = cfactor
    adj = cfactorAll.T

    # Getting the inverse
    inv = adj / det
    #returning the inverse
    return inv




inv(a)

```

##### Problem 2:
Complete the code below to display an RGB color image ensuring the following steps:


*   Click [here](https://drive.google.com/file/d/15VT8TcJj6mF8hFFbJsXQ18hsbcHzt1J1/view?usp=sharing) to download the "data.npz" file and upload it in the current working directory.
*   Use the `np.linalg.inv` method to find the inverse matrix of `R`, `G` and `B` matrices.
*   Combine the inverse `R`, `G` and `B` matrices (each of shape $2000 \times 2000$) into a single matrix of shape $2000 \times 2000 \times 3$ such that each of the three matrices represent the RGB color channels [Hint: Create a zero matrix of `dtype = "uint8"` with the shape equal to the shape of `R`, `G` or `B`, and store their inverse matrix in `img` variable using appropriate indexing.]
*   Display the img variable as an image using `plt.imshow` method.



```python
import numpy as np
import matplotlib.pyplot as plt


data = np.load('data.npz')

R = data['R']
G = data['G']
B = data['B']
 
#FINDING THE INVERSE OF MATRICES
R_inverse = np.linalg.inv(R)
G_inverse = np.linalg.inv(G)
B_inverse = np.linalg.inv(B)

#ZERO MATRIX
img = np.zeros((2000,2000,3), dtype="uint8")

#IMAGE VARIABLE STORE
img[:,:,0] = R_inverse
img[:,:,1] = G_inverse
img[:,:,2] = B_inverse
 
#PLOTTING 
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# importing necessary packages
import numpy as np
import matplotlib.pyplot as plt

xR = 0
x = 0.05

#Here we are  going to calculate the  fourth ORDER TAYLOR APPROX FOR e^(2x)
def fZeroPrime(x):
  return np.exp(2*x)

def fOnePrime(x):
  return 2*fZeroPrime(x) 

def fTwoPrime(x):
  return 4*fZeroPrime(x)

def fThreePrime(x):
  return 8*fZeroPrime(x)

def fFourPrime(x):
  return 16*fZeroPrime(x) 

def taylorFourthOrder(x):
  return fZeroPrime(xR) + (fOnePrime(xR)*(x-xR)**1)/1 + (fTwoPrime(xR)*(x-xR)**2)/2 + (fThreePrime(xR)*(x-xR)**3)/6 + (fFourPrime(xR)*(x-xR)**4)/24

print("The fourth order Polynomial for e^(2x) is : \n")
print(taylorFourthOrder(x))  
print("\n\n")

#CALCULATING 4TH ORDER TAYLOR APPROX FOR e^(5x)
def fZero5e(x):
  return np.exp(5*x)

def fOne5e(x):
  return 5*fZero5e(x) 

def fTwo5e(x):
  return 25*fZero5e(x)

def fThree5e(x):
  return 125*fZero5e(x)

def fFour5e(x):
  return 625*fZero5e(x) 

def taylorFourthOrder5e(x):
  return fZero5e(xR) + (fOne5e(xR)*(x-xR)**1)/1 + (fTwo5e(xR)*(x-xR)**2)/2 + (fThree5e(xR)*(x-xR)**3)/6 + (fFour5e(xR)*(x-xR)**4)/24

print("The fourth order polynomial for e^(5x) is:\n")
print(taylorFourthOrder5e(x)) 
print("\n\n")  


#CALCULATING fourth ORDER TAYLOR APPROX FOR e^(-3x)
def fZero3e(x):
  return np.exp(-3*x)

def fOne3e(x):
  return -3*fZero3e(x) 

def fTwo3e(x):
  return 9*fZero3e(x)

def fThree3e(x):
  return -27*fZero3e(x)

def fFour3e(x):
  return 81*fZero3e(x)  

def taylorFourthOrder_3e(x):
  return fZero3e(xR) + (fOne3e(xR)*(x-xR)**1)/1 + (fTwo3e(xR)*(x-xR)**2)/2 + (fThree3e(xR)*(x-xR)**3)/6 + (fFour3e(xR)*(x-xR)**4)/24

print("The Fourth order polynomial for e^(-3x) is:\n")
print(taylorFourthOrder_3e(x))
print("\n\n") 

multiplication =  taylorFourthOrder5e(x) * taylorFourthOrder_3e(x)
actual = np.exp(2*x)
print("So the taylor approximation of e^(5x) * e^(-3x):\n\n")
print(multiplication)

#PLOTTING OF POINTS
t = np.linspace(0, 5, 10)
y = fZeroPrime(np.linspace(0, 1, 10))
plt.plot(t, y, color = "k")

plt.plot(taylorFourthOrder(x), 'yo')
plt.plot(actual, 'go')
plt.plot(multiplication, 'ro')
plt.show()
```
